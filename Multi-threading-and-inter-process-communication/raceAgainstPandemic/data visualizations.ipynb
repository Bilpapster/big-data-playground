```python
import matplotlib.pyplot as plt
import pandas as pd

FILE_NAME = 'data.csv'
```

```python
data = pd.read_csv(FILE_NAME)
data['time'] = data['time'] / 1000.0
data.head()
```

```python
# plot patients in ICU and out of ICU

x = data['time']
color_in = 'orange'
color_out = 'red'
plt.figure(figsize=(20,10))
plt.plot(x, data['in'], color=color_in, label='patients in ICU')
plt.plot(x, data['out'], color=color_out, label='patients waiting for ICU')
plt.legend()
plt.fill_between(x, data['in'], data['out'], alpha=0.1, color=color_out, where=data['out'] > data['in'])
plt.fill_between(x, data['out'], data['in'], alpha=0.1, color=color_in, where=data['in'] > data['out'])
plt.show()
```

```python
# plot cumulative treatments vs cumulative infections

x = data['time']
color_treatments = 'green'
color_infections = 'red'
plt.figure(figsize=(20,10))
plt.plot(x, data['treatments'], color=color_treatments, label='cumulative treatments')
plt.plot(x, data['infections'], color=color_infections, label='cumulative infections')
plt.legend()
# plt.fill_between(x, data['treatments'], data['infections'], alpha=0.1, color=color_infections, where=data['infections'] > data['treatments'])
# plt.fill_between(x, data['infections'], data['treatments'], alpha=0.1, color=color_treatments, where=data['treatments'] > data['infections'])
plt.show()
```
